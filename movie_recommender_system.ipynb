# Downloading data from git repo


```python
import urllib.request
import zipfile
import os

def download_and_extract_zip(url, csv_file_name):
    # Path to save the downloaded zip file
    zip_file_path = "/content/downloaded_file.zip"

    # Download the zip file
    urllib.request.urlretrieve(url, zip_file_path)

    # Extract the zip file
    with zipfile.ZipFile(zip_file_path, "r") as zip_ref:
        zip_ref.extractall("/content/")

    # Path to the extracted CSV file
    extracted_csv_path = f"/content/{csv_file_name}.csv"

    # Rename the extracted file
    extracted_files = os.listdir("/content/")
    for file in extracted_files:
        if file.endswith(".csv"):
            old_file_path = os.path.join("/content/", file)
            new_file_path = os.path.join("/content/", f"{csv_file_name}.csv")
            os.rename(old_file_path, new_file_path)
            break

    # Remove the downloaded zip file
    os.remove(zip_file_path)

    return extracted_csv_path
```

```python
credits_url = "https://github.com/uniyalmani/dataset/raw/main/tmdb_5000_credits.csv.zip"
movie_url = "https://github.com/uniyalmani/dataset/raw/main/tmdb_5000_movies.csv.zip"
credit_file_name = "tmdb_5000_credits"
movie_file_name = "tmdb_5000_movie"


credits_csv_file_path = download_and_extract_zip(credits_url, credit_file_name)
movies_csv_file_path = download_and_extract_zip(movie_url, movie_file_name)

print(credits_csv_file_path)
print(movies_csv_file_path)
```

```python
movies = pd.read_csv(movies_csv_file_path)
credits = pd.read_csv(credits_csv_file_path)

movies.head(1)
```

```python
credits.head(1)
```

```python
movies = movies.merge(credits, on = "title")
movies.head(1)
```

```python
"""
important columns
genres
id
keywords
title
overview
cast
crew
"""
movies = movies[["movie_id", "title", "overview", "genres", "keywords", "cast", "crew"]]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
import ast
def formate_col(obj, col_name):
  lst = []
  #ast.literal_eval() for converting str to other data structure
  for info in ast.literal_eval(obj): 
    lst.append(info[col_name])
  return lst
```

```python
movies["genres"] = movies["genres"].apply(lambda x: formate_col(x, "name")) #for passing multiple args to function inside apply use lambda func
```

```python
movies["keywords"] = movies["keywords"].apply(lambda x: formate_col(x, "name"))
```

```python
def formate_col_with_limit(obj, col_name, limit):
  lst = []
  #ast.literal_eval() for converting str to other data structure
  count = 0
  for info in ast.literal_eval(obj): 
    if count >= limit:
      break
    lst.append(info[col_name])
    count += 1
  return lst
```

```python
movies["cast"] = movies["cast"].apply(lambda x: formate_col_with_limit(x, "name", 3))
```

```python
movies.head()
```

```python
def fetch_director(obj, col_name):
  lst = []
  #ast.literal_eval() for converting str to other data structure
  for info in ast.literal_eval(obj): 
    if info["job"] == "Director":
      lst.append(info[col_name])
      break
  return lst
```

```python
movies["crew"] = movies["crew"].apply(lambda x: fetch_director(x, "name"))
```

```python
movies.head()
```

```python
movies["overview"] = movies["overview"].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
movies["genres"] = movies["genres"].apply(lambda x: [i.replace(" ", "") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x: [i.replace(" ", "") for i in x])
movies["cast"] = movies["cast"].apply(lambda x: [i.replace(" ", "") for i in x])
movies["crew"] = movies["crew"].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies["tags"] =  movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
```

```python
movies.head()
```

```python
new_df = movies[["movie_id", "title", "tags"]]
```

```python
new_df.head()
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x: x.lower())
```

```python
import nltk #nlp (for Stemming)

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
  y = []

  for word in text.split():
    y.append(ps.stem(word))

  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
# we need to change tags into a vector (bag of words, tfidf , word to vec)
"""
bag of words -> concatinate all  .  final_tag =  tags --> tag1 + tag2 ....
calculate freq of each word and select top n words with higher freq
and create vector using freq -> 
              word1  word2  word3 ......  wordn  (check freq of this word in evey tag of movies and create vector from it)
for movie1 ->  k1      k2    k3             kn --> vector for movie1
    movie2 ->  x1       x2   x3             xn  --> vector for movie2

we can use sklearn

and apply Stemming(Stemming is a technique used to extract the base form of the words by removing affixes from them)
and calculate cosine distance
"""

from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
vectors = cv.fit_transform(new_df["tags"]).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distance = similarity[movie_index]
  movies_list = sorted(list(enumerate(distance)), reverse = True, key = lambda x: x[1])[1: 6]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
   
  return 
```

```python
recommend("Avatar")
```
